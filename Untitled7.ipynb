```python
import pandas as pd

url = 'https://github.com/mattharrison/datasets/raw/refs/heads/master/data/ames-housing-dataset.zip'

df = pd.read_csv(url, compression='zip')
```

```python
df['TotalBsmtSF'].corr(df['SalePrice'])
```

```python
df['SalePrice'].mean()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['YearBuilt'], df['SalePrice'], alpha=0.5)
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.title('Sale Price vs. Year Built')
plt.show()
```

```python
plt.scatter(df['Year Built'], df['SalePrice'], alpha=0.5)
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.title('Sale Price vs Year Built')
plt.show()
```

```python
plt.scatter(df['TotalBsmtSF'], df['SalePrice'], alpha=0.5)
```

```python
plt.scatter(df['Total Bsmt SF'], df['SalePrice'], alpha=0.5)
plt.xlabel('Total Basement SF')
plt.ylabel('Sale Price')
plt.title('Sale Price vs Basement Size')
plt.show()
```

```python
df.groupby('Year Built')['SalePrice'].mean().plot(kind='line')
plt.title('Average Sale Price Over Time')
plt.ylabel('Sale Price')
plt.show()
```
